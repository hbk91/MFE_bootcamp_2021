# Central Limit Theorem

```python
# Importing required libraries
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(10)
```

```python
# Creating a population
population = np.concatenate((np.random.chisquare(10,10000),
                             np.random.chisquare(30,10000),
                             np.random.exponential(10,10000)))

plt.hist(population,bins=250)
plt.title("Distribution of Population");plt.show()
print("Population Mean", np.mean(population))
```

```python
# Generating sample mean
sample_size = [10,100,1000]
simulations = 10000

sample_mean_10 = []
sample_mean_100 = []
sample_mean_1000 = []

for j in sample_size:
    for i in range(simulations):
        if j == 10 :
            sample = np.random.choice(population,j)
            sample_mean_10.append(np.mean(sample))
        elif j == 100 : 
            sample = np.random.choice(population,j)
            sample_mean_100.append(np.mean(sample))
        else : 
            sample = np.random.choice(population,j)
            sample_mean_1000.append(np.mean(sample))
    
# Plotting histogram of sample means
plt.hist(sample_mean_10,alpha=0.75,bins=int(simulations/10),label="ss=10")
plt.hist(sample_mean_100,alpha=0.5,bins=int(simulations/10),label="ss=100")
plt.hist(sample_mean_1000,bins=int(simulations/10),label="ss=1000")
plt.legend()
plt.title("Distribution of Sample Mean (Varying Sample Sizes)");plt.rcParams["figure.figsize"] = (12,5)
plt.show()
```

```python
# Generating sample mean
sample_size = 1000
simulations = [100,1000,10000]

sample_mean_100 = []
sample_mean_1000 = []
sample_mean_10000 = []

for i in simulations:
    for j in range(i):
        sample = np.random.choice(population,sample_size)
        if i == 100: sample_mean_100.append(np.mean(sample))
        if i == 1000: sample_mean_1000.append(np.mean(sample))
        if i == 10000: sample_mean_10000.append(np.mean(sample))

# Plotting histogram of sample means
plt.hist(sample_mean_100,bins=25,label="n=100",color='k')
plt.hist(sample_mean_1000,bins=100,alpha=0.35,label="n=1000",color='b')
plt.hist(sample_mean_10000,bins=250,alpha=0.25,label="n=10000",color='orange')
plt.legend()
plt.title("Distribution of Sample Mean (Varying Simulation Sizes)");plt.rcParams["figure.figsize"] = (12,5)
plt.show()
```
