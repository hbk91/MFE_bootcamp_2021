```python
import numpy as np
import matplotlib.pyplot as plt
from math import factorial
```

# Binomial Distribution 

```python
# Distribution of binomial distribution

# Initializing parameters
n = 100
p1 = 0.2
p2 = 0.5
p3 = 0.8

# Plotting PMF for different probabilities
pmf1 = [factorial(n)/(factorial(k)*factorial(n-k))*(p1**k)*((1-p1)**(n-k))  for k in range(n+1)]
pmf2 = [factorial(n)/(factorial(k)*factorial(n-k))*(p2**k)*((1-p2)**(n-k))  for k in range(n+1)]
pmf3 = [factorial(n)/(factorial(k)*factorial(n-k))*(p3**k)*((1-p3)**(n-k))  for k in range(n+1)]

plt.plot(range(n+1),pmf1,label="n=100,p=0.2",color='k')
plt.plot(range(n+1),pmf2,label="n=100,p=0.5",color='b')
plt.plot(range(n+1),pmf3,label="n=100,p=0.8",color='r')
plt.legend(loc=5);plt.show()
```

```python
# Distribution of binomial distribution

# Initializing parameters
n1 = 10
n2 = 50
n3 = 100
p = 0.3

# Plotting PMF for different values of n
pmf1 = [factorial(n1)/(factorial(k)*factorial(n1-k))*(p**k)*((1-p)**(n1-k))  for k in range(n1+1)]
pmf2 = [factorial(n2)/(factorial(k)*factorial(n2-k))*(p**k)*((1-p)**(n2-k))  for k in range(n2+1)]
pmf3 = [factorial(n3)/(factorial(k)*factorial(n3-k))*(p**k)*((1-p)**(n3-k))  for k in range(n3+1)]

plt.plot(range(n1+1),pmf1,label="n=10,p=0.3",color='k',lw=4)
plt.plot(range(n2+1),pmf2,label="n=100,p=0.3",color='b',lw=2.5)
plt.plot(range(n3+1),pmf3,label="n=1000,p=0.3",color='red',lw=1.5)
plt.legend();plt.show()
```

```python
# Comparing simulations vs actual PMF for binomial distribution

# Initializing variables
n = 100
p = 0.3
simulations = 10000

# Computing theoretical PMF
pmf_formula = [factorial(n)/(factorial(k)*factorial(n-k))*(p**k)*((1-p)**(n-k))  for k in range(n+1)]

# Simulating PMFs
binom_simulation = np.random.binomial(n=n,p=p,size=simulations)

# Counting unique values
x,count = np.unique(binom_simulation,return_counts=True)
pmf_simulation = count/binom_simulation.size

# Plotting the results
plt.plot(x,pmf_simulation,label='PMF Simulation',lw=2)
plt.plot(range(n+1),pmf_formula,label='PMF Formula',color='k',alpha=0.5) # Overlaying the actual PMF
plt.legend();plt.show()
```

# Poisson Distribution

```python
# Distribution of binomial distribution

# Initializing parameters
n = 100
λ1 = 1
λ2 = 5
λ3 = 50


# Plotting PMF for different probabilities
pmf1 = [np.exp(-λ1)*(λ1**x)/factorial(x)  for x in range(n+1)]
pmf2 = [np.exp(-λ2)*(λ2**x)/factorial(x)  for x in range(n+1)]
pmf3 = [np.exp(-λ3)*(λ3**x)/factorial(x)  for x in range(n+1)]

plt.plot(range(n+1),pmf1,label="λ = 1",color='k')
plt.plot(range(n+1),pmf2,label="λ = 5",color='b')
plt.plot(range(n+1),pmf3,label="λ = 50",color='r')
plt.legend(loc=5);plt.show()
```

```python
# Comparing simulations vs actual PMF for binomial distribution

# Initializing variables
λ = 20
simulations = 10000

# Computing theoretical PMF
pmf_formula = [np.exp(-λ)*(λ**x)/factorial(x)  for x in range(n+1)]

# Simulating PMFs
poisson_simulation = np.random.poisson(lam=λ,size=simulations)

# Counting unique values
x,count = np.unique(poisson_simulation,return_counts=True)
pmf_simulation = count/poisson_simulation.size

# Plotting the results
plt.plot(x,pmf_simulation,label='PMF Simulation',lw=2)
plt.plot(range(n+1),pmf_formula,label='PMF Formula',color='k',alpha=0.5) # Overlaying the actual PMF
plt.legend();plt.show()
```

# Binomial to Poisson Convergence

```python
# Setting expected values equal for poisson and binomial

# Initializing n and p 
n = 100
p = 0.01 # making p smaller or n bigger leads to convergence
λ = n*p

# Computing PMF of binomial and poisson using their PMFs
pmf_binomial = [factorial(n)/(factorial(k)*factorial(n-k))*(p**k)*((1-p)**(n-k))  for k in range(n+1)]
pmf_poisson = [np.exp(-λ)*(λ**x)/factorial(x)  for x in range(n+1)]

# Plotting the two PMFs
plt.plot(range(n+1),pmf_binomial,label='PMF Binomial',color='k')
plt.plot(range(n+1),pmf_poisson,label='PMF Poisson',color='b')
plt.legend();plt.show()
```

# Memoryless Property of Poisson and Exponential Distribution

$$P(X>t+s|x>s) = \frac{P(X>t+s,X>s)}{P(X>s)} = \frac{P(X>t+s)}{P(X>s)} = \frac{1-F_X(t+s)}{1-F_X(s)} = P(X>t) = 1-F_X(t)$$

```python
# Calculating the exponential distribution CDF
λ = 0.5
n = 20
cdf_exponential = [1-np.exp(-λ*x) for x in range(n+1)]

# Looping over all variables (assuming t = 2, s = 1)
t = 2
s = 1
p_ts = []
p_t = []
p_s = []
while t+s<len(cdf_exponential):
    p_ts.append(1-cdf_exponential[t+s])
    p_t.append(1-cdf_exponential[t])
    p_s.append(1-cdf_exponential[s])
    s = s+1
    t=t+1

# Plotting the results
```
